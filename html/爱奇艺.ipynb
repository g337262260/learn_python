```python
def get_text_link_from_sel(sel):
    mylist = []
    try:
        results = r.html.find(sel)
        for result in results:
            if '主演' in result.text:
                p = result.text.index('主演')
                mylist.append(result.text[p:])
            else:
                mylist.append('未知')
        return mylist
    except:
        return None
```

```python
print(get_text_link_from_sel(sel))
```
