```python
from requests_html import HTMLSession
```

```python
session = HTMLSession()
```

```python
url = 'http://www.xiaoheju.net/player/13286-1-1.html'
```

```python
r = session.get(url)
```

```python
print(r.html.text)
```

```python
r.html.links
```

```python
r.html.absolute_links
```

```python
sel = '#content > div > div.article > div:nth-child(2) > div > table > tbody > tr > td > a'
```

```python
results = r.html.find(sel)
```

```python
results
```

```python
results[0].text
```

```python
results[0].absolute_links
```

```python
list(results[0].absolute_links)[0]
```

```python
def get_text_link_from_sel(sel):
    mylist = []
    try:
        results = r.html.find(sel)
        for result in results:
            mytext = result.text
            mylink = list(result.absolute_links)[0]
            mylist.append((mytext, mylink))
        return mylist
    except:
        return None
```

```python
print(get_text_link_from_sel(sel))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(get_text_link_from_sel(sel))
```

```python
df.columns = ['book', 'link']
```
